# **Decisões**

Os comandos para tomada de decisão são muito úteis caso queiramos seguir caminhos diferentes num algoritmo em função de situações diferentes.

Captura de tela de 2021-02-24 19-10-42.png

Captura de tela de 2021-02-24 19-11-08.png

Em Python temos as seguintes estruturas:


1.   Execução de comandos caso uma condição seja verdadeira:
```
if condição:
        comandos executados caso a condição seja verdadeira 
        .....
```
2.   Execução de comandos caso uma condição seja verdadeira e de outros comandos caso seja falsa:
```
if condição:
        comandos executados caso a condição seja verdadeira 
        ....
else:
        comandos executados caso a condição seja falsa 
        .....
```
3.   Execução de comandos associados a condições mutuamente exclusivas (há apenas uma verdadeira entre as várias condições): 
```
if condição1:
        comandos executados caso a condição1 seja verdadeira 
        .....
elif condição2:
        comandos executados caso a condição2 seja verdadeira 
        .....
elif condiçãoN:
        comandos executados caso a condiçãoN seja verdadeira 
        .....
else:
        comandos executados caso todas as condições sejam falsas 
        .....
```

**O que pode ser uma condição?**

Os operadores relacionais são os seguintes:
*  ==  →  comparação de igualdade
*  !=  →  comparação de diferença
*  \>  →  maior que
*  \>=  →  maior ou igual a
*  \<  →  menor que
*  \<=  →  menor ou igual a

**Operadores lógicos:**

Há um modo de compor condições mais complexas com o uso de operadores lógicos (E, OU e NÃO):

*   E (and): também chamado de conjunção, tem um paralelo na intersecção de conjuntos.
```
if condição1 and condição2:
        comandos executados caso AMBAS as condições sejam verdadeiras
```

Captura de tela de 2021-02-26 09-14-58.png

*   OU (or): também chamado de disjunção, tem um paralelo na união de conjuntos.
```
if condição1 or condição2:
        comandos executados caso QUALQUER uma das condições seja verdadeira
```

Captura de tela de 2021-02-26 09-16-32.png

*   NÃO (not): também chamado de negação, tem um paralelo no complemento de conjuntos.
```
if not condição1:
        comandos executados caso a condição NÃO seja verdadeira
```

5) Cálculo do Índice de Massa Corporal (IMC):

$$
IMC = \frac{peso}{altura^2} \\
$$


Captura de tela de 2021-02-24 18-15-57.png

[Referência](https://www.endocrino.org.br/teste-seu-imc/)

Captura de tela de 2021-08-13 10-44-36.png

Captura de tela de 2021-02-24 18-20-59.png

[Link do Blockly](https://blockly-demo.appspot.com/static/demos/code/index.html?lang=pt-br#qj7kam)

````
peso = float(input('Entre com seu peso (em kg): '))
altura = float(input('Entre com sua altura (em m): '))
imc = peso / altura**2  # ** corresponde à potenciação
print('Seu IMC vale', imc)
# observe agora a formatação que limita o número de casas decimais após o ponto
print(f'Seu IMC vale {imc:.2f}')
if imc < 18.5:
  print('Magro de ruim....')
elif imc < 25:
  print('Normal....')
elif imc < 30:
  print('Fortinho....')
elif imc < 40:
  print('Fortão....')
else:
  print('Bariátrica já!!!!')
````
